Implement geometric transformations on an image, including 
translation, rotation, and scaling. Display the original image 
alongside the transformed images for comparison.

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread("image.png")  # Replace with your image path
if img is None:
    raise FileNotFoundError("Image not found. Check the file path.")

# Get dimensions and center
h, w = img.shape[:2]
center = (w // 2, h // 2)

# Translation: shift right 50px and down 30px
M_trans = np.float32([[1, 0, 50], [0, 1, 30]])
translated = cv2.warpAffine(img, M_trans, (w, h))

# Rotation: rotate 45 degrees around center
M_rot = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M_rot, (w, h))

# Scaling: scale 1.5x and resize back to original size for display
scaled = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
scaled_resized = cv2.resize(scaled, (w, h))

# Combine all images horizontally
combined = np.hstack([img, translated, rotated, scaled_resized])

# Display result
cv2.imshow("Original | Translated | Rotated | Scaled", combined)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("image.png")  # Replace with your image path
if img is None:
    raise FileNotFoundError("Image not found. Check the file path.")

# Convert from BGR (OpenCV) to RGB (Matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get dimensions and center
h, w = img.shape[:2]
center = (w // 2, h // 2)

# Translation: shift right 50px and down 30px
M_trans = np.float32([[1, 0, 50], [0, 1, 30]])
translated = cv2.warpAffine(img, M_trans, (w, h))
translated_rgb = cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)

# Rotation: rotate 45 degrees around center
M_rot = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M_rot, (w, h))
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)

# Scaling: scale 1.5x and resize back to original size for display
scaled = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
scaled_resized = cv2.resize(scaled, (w, h))
scaled_rgb = cv2.cvtColor(scaled_resized, cv2.COLOR_BGR2RGB)

# Plotting using matplotlib
titles = ['Original', 'Translated', 'Rotated', 'Scaled']
images = [img_rgb, translated_rgb, rotated_rgb, scaled_rgb]

plt.figure(figsize=(12, 4))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
