1.  Basic Image Handling and Processing Operations 


```python
import cv2
import matplotlib.pyplot as plt

# Load image (replace with your image path)
image_path = "image.png"
original = cv2.imread(image_path)

if original is None:
    raise FileNotFoundError(f"Could not load image from {image_path}")

# Convert BGR to RGB for proper Matplotlib display
original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

# Basic image operations
grayscale = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(original, (0,0), fx=0.5, fy=0.5)  # Half size
edges = cv2.Canny(original, 100, 200)

# Create figure with subplots
plt.figure(figsize=(12, 8))

# Display original image
plt.subplot(2, 2, 1)
plt.imshow(original_rgb)
plt.title("Original Image")
plt.axis('off')

# Display grayscale
plt.subplot(2, 2, 2)
plt.imshow(grayscale, cmap='gray')
plt.title("Grayscale")
plt.axis('off')

# Display resized image (convert to RGB first)
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized")
plt.axis('off')

# Display edges
plt.subplot(2, 2, 4)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')

# Adjust layout and show plot
plt.tight_layout()
plt.show()

```

```python
import cv2

def main():
    # Load image (replace with your image path)
    image_path = "input_image.jpg"
    original = cv2.imread(image_path)
    
    if original is None:
        print(f"Error: Could not load image from {image_path}")
        return

    # Basic image operations
    grayscale = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(original, (0,0), fx=0.5, fy=0.5)  # Half size
    edges = cv2.Canny(original, threshold1=100, threshold2=200)

    # Display results
    cv2.imshow("Original Image", original)
    cv2.imshow("Grayscale Conversion", grayscale)
    cv2.imshow("Resized Image", resized)
    cv2.imshow("Edge Detection", edges)
    
    # Save processed images
    cv2.imwrite("grayscale_output.jpg", grayscale)
    cv2.imwrite("resized_output.jpg", resized)
    cv2.imwrite("edge_detection_output.jpg", edges)

    # Wait for key press and clean up
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
