Convert an image from RGB to different color spaces (e.g., HSV, 
LAB, YCbCr). Display the original image and each converted
image side by side, explaining the significance of each color space. 


```python
import cv2
import numpy as np

# Load image
img = cv2.imread('image.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

# Convert to different color spaces
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
ycbcr = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)  # Note: YCrCb is YCbCr in OpenCV

# Split channels for each color space
hsv_channels = cv2.split(hsv)
lab_channels = cv2.split(lab)
ycbcr_channels = cv2.split(ycbcr)

# Create a blank image for channel display
blank = np.zeros_like(img[:,:,0])

# Display all images
cv2.imshow('Original (RGB)', img_rgb)
cv2.imshow('HSV - Hue', hsv_channels[0])
cv2.imshow('HSV - Saturation', hsv_channels[1])
cv2.imshow('HSV - Value', hsv_channels[2])
cv2.imshow('LAB - L*', lab_channels[0])
cv2.imshow('LAB - a*', lab_channels[1])
cv2.imshow('LAB - b*', lab_channels[2])
cv2.imshow('YCbCr - Y', ycbcr_channels[0])
cv2.imshow('YCbCr - Cb', ycbcr_channels[1])
cv2.imshow('YCbCr - Cr', ycbcr_channels[2])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('image.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to color spaces
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
ycbcr = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)  # OpenCV uses YCrCb for YCbCr

# Split channels
hsv_ch = cv2.split(hsv)
lab_ch = cv2.split(lab)
ycbcr_ch = cv2.split(ycbcr)

# Titles and channel list
titles = [
    'Original (RGB)',
    'HSV - Hue', 'HSV - Saturation', 'HSV - Value',
    'LAB - L*', 'LAB - a*', 'LAB - b*',
    'YCbCr - Y', 'YCbCr - Cb', 'YCbCr - Cr'
]
images = [img_rgb] + list(hsv_ch) + list(lab_ch) + list(ycbcr_ch)
cmaps = [None] + ['gray'] * 9  # Use RGB for original, grayscale for channels

# Plot
plt.figure(figsize=(15, 8))
for i, (img, title, cmap) in enumerate(zip(images, titles, cmaps)):
    plt.subplot(2, 5, i + 1)
    plt.imshow(img if cmap is None else img, cmap=cmap)
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
