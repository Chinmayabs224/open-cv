Apply a Gaussian filter to an image to reduce noise. Display the 
original and filtered images side by side for comparison

```python
import cv2
import numpy as np

img = cv2.imread('image.png')  # Load image

# Define kernel sizes (progressively increasing)
kernel_sizes = [(3, 3), (9, 9), (15, 15), (25, 25)]  # Increasing blur strength

# Apply Gaussian blur for each kernel size
blurred_images = [img]  # Start with original image
for ksize in kernel_sizes:
    blurred = cv2.GaussianBlur(img, ksize, 0)
    blurred_images.append(blurred)

# Combine all images horizontally
combined = np.hstack(blurred_images)

# Display results
cv2.imshow('Original | Increasing Gaussian Blur', combined)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load and convert image to RGB
img = cv2.imread('image.png')  # Load image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define kernel sizes (increasing blur)
kernel_sizes = [(3, 3), (9, 9), (15, 15), (25, 25)]

# Apply Gaussian blur
blurred_images = [img_rgb] + [cv2.GaussianBlur(img_rgb, k, 0) for k in kernel_sizes]
titles = ['Original'] + [f'Blur {k[0]}x{k[1]}' for k in kernel_sizes]

# Plot images side-by-side
plt.figure(figsize=(15, 5))
for i, (img, title) in enumerate(zip(blurred_images, titles)):
    plt.subplot(1, len(blurred_images), i + 1)
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
